```python
#E-COMMERCE Sales Analysis using Numpy
import numpy as np
np.random.seed(42)
sales = np.random.randint(1000, 20000, 30)
print("Daily Sales")
print(sales)
print("\nTotal Sales:", sales.sum())
print("Average Sales:", sales.mean())
print("Highest Sale:", sales.max())
print("Lowest Sale:", sales.min())
best_day = np.argmax(sales)+1
worst_day = np.argmin(sales)+1
print("Best Sales Day:", best_day)
print("Worst Sales Day:", worst_day)
print("\nDays Above Average")
print(np.where(sales > sales.mean())[0]+1)
```
