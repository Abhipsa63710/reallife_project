```python
#Hospital patient analytics
import numpy as np
waiting_time = np.random.randint(5,120,200)
print("Average Waiting Time:",waiting_time.mean())
print("Longest Waiting:",waiting_time.max())
print("Patients Waiting > 60 minutes")
print(waiting_time[waiting_time>60])
print("Total Patients:",waiting_time.size)
```
